## How to convert a 1D array into a 2D array 
Adding new axis to an array

(    ) 

<b>newaxis</b>


`np.newaxis` enables us to increase the dimensions of our existing array.


1D array will becomes a 2D array 

2D array will become a 3D array

and so on.

```python
x = np.array([1, 2, 3, 4, 5, 6])
x.shape
```

```python
# Let's use np.newaxis to add a new axis
x2 = x[np.newaxis]
x2.shape
```

```python
x2
```

\_()_/ <b>NOTE</b>


We can explicitly convert a 1D array with either a row vector or a column vector using `np.newaxis`. 


We can convert a 1D array to a row vector by inserting an axis along the first dimension.

```python
row_vector = x[np.newaxis, :]
row_vector.shape
```

```python
row_vector
```

```python
col_vector = x[:, np.newaxis]
col_vector.shape
```

```python
col_vector
```

# How to get unique items and counts

```python
# find the unique elements in an array easily 
```

```python
mixed_array = np.array([5,5,11,11,2,3,4,8,14,14,15,5])
unique_vals = np.unique(mixed_array)

unique_vals
```

```python
# To get the indices of unique values in a NumPy array 
# (an array of first index positions of unique values in the array)

unique_vals_idx = np.unique(mixed_array, return_index=True)
unique_vals_idx
```

```python
# You can pass the return_counts argument in np.unique() 
# along with your array to get the frequency count of unique values in a NumPy array.

unique_vals_idx = np.unique(mixed_array, return_counts=True)
unique_vals_idx
```

```python
# More interesting graphs using NumPy arrays - remember linspace is evenly spaced numbers

x = np.linspace(0, 15, 20)
y = np.linspace(0, 25, 20)
plt.plot(x, y, 'red') # line
plt.plot(x, y, 'o')      # dots
```

```python


from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)
X = np.arange(-5, 5, 0.15)
Y = np.arange(-5, 5, 0.15)
# This results in array from 67 col -> 67 row x 67 col array
X, Y = np.meshgrid(X, Y)
# X.shape
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis')
```
