# Challenge 2

<div class="alert alert-block alert-info">
Use company sales data to fetch all product sales figures and show them using a multiline plot
</div>

- Display the number of units sold per month for each product using multiline plots. 
- (i.e., Separate Plotline for each product ).
 
 
 ### 20 min START NOW

```python
plt.subplots(figsize=(30,18)) 

x = df['month_number']
y1 = df['facecream']
y2 = df['facewash']
y3 = df['toothpaste']
y4 = df['bathingsoap']
y5= df['shampoo']
y6 = df['moisturizer']

plt.bar(x, y1, label='facecream', linewidth=5, color='r')
plt.bar(x, y2, label='facewash', linewidth=10, color='black')
plt.plot(x, y3, 'g', label='toothpaste', linewidth=5)
plt.plot(x, y4, 'y', label='bathingsoap', color='coral', linewidth=5)
plt.plot(x, y5, label='shampoo', linewidth=5, color='plum')
plt.plot(x, y6,  label='moisturizer', linewidth=5, color = 'yellow')

plt.title('Product Sales by Month', fontsize=40)

plt.xticks(fontsize=30)
plt.yticks(fontsize=30)

plt.xlabel('Month',fontsize=50)
plt.ylabel('Product Sales',fontsize=50)

plt.legend(loc=2, fontsize=30)
plt.grid()
```

```python
import matplotlib as plt
import matplotlib.pyplot as plt

plt.subplots(figsize=(30,18)) 

x = df['month_number']
y1 = df['facecream']
y2 = df['facewash']
y3 = df['toothpaste']
y4 = df['bathingsoap']
y5= df['shampoo']
y6 = df['moisturizer']

plt.plot(x, y1, '--',label='facecream', linewidth=5, c='r')
plt.plot(x, y2, label='facewash', linewidth=10, color='black')
plt.plot(x, y3, 'g', label='toothpaste', linewidth=5)
plt.plot(x, y4, 'y', label='bathingsoap', color='coral', linewidth=5)
plt.plot(x, y5, label='shampoo', linewidth=5, color='plum')
plt.plot(x, y6, '--', label='moisturizer', linewidth=5, color = 'yellow')

plt.title('Product Sales by Month', fontsize=40)

plt.xticks(fontsize=30)
plt.yticks(fontsize=30)

plt.xlabel('Month',fontsize=50)
plt.ylabel('Product Sales',fontsize=50)

plt.legend(loc=2, fontsize=30)
plt.grid()
```
